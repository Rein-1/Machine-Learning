## Machine Learning

### Setup

```python
from sklearn.metrics import accuracy_score

# Training the model
def learning_curve(model, epochs=10):
    train_accuracies = []
    test_accuracies = []

    for epoch in range(epochs):
        model.fit(X_train, y_train)

        # Training accuracy
        y_train_pred = model.predict(X_train)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        train_accuracies.append(train_accuracy)

        # Test accuracy
        y_test_pred = model.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        test_accuracies.append(test_accuracy)

    # Plotting the learning curve
    epochs_range = range(1, epochs + 1)

    plt.figure(figsize=(6, 5))
    plt.plot(epochs_range, train_accuracies, label='Training Accuracy', marker='o')
    plt.plot(epochs_range, test_accuracies, label='Test Accuracy', marker='o')
    plt.title('Learning Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.xlim(0, epochs+1)
    plt.ylim(0, 1)
    plt.legend()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve, roc_auc_score


def plot_roc_curve(*models):
    plt.figure(figsize=(6, 5))
    colors = ['blue', 'green', 'red', 'darkturquoise', 'darkgoldenrod', 'darkorange']

    for model in models:
        # Get the probabilities for each of the two categories
        name = model[0]
        y_prob = model[1].predict_proba(X_test)[:,1]

        auroc = roc_auc_score(y_test, y_prob)
        fpr, tpr, thresholds = roc_curve(y_test, y_prob)
        plt.plot(fpr, tpr, color=colors.pop(), lw=0.5, label=f'{name} (AUC = {auroc:.4f})')

    # It is customary to draw a diagonal dotted line in ROC plots.
    # This is to indicate completely random prediction. Deviation from this
    # dotted line towards the upper left corner signifies the power of the model.
    plt.plot([0, 1], [0, 1], ls='--', color='black', lw=.3)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.xlim(-.01, 1.01)
    plt.ylim(-.01, 1.01)
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.grid(True)

def plot_precision_recall_curve(*models):
    plt.figure(figsize=(6, 5))
    colors = ['blue', 'green', 'red', 'darkturquoise', 'darkgoldenrod', 'darkorange']

    for model in models:
        # Plot the precision-recall curve
        y_prob = model[1].predict_proba(X_test)
        y_pred = model[1].predict(X_test)
        name = model[0]

        precision, recall, _ = precision_recall_curve(y_test, y_prob[:,1])
        f1score = f1_score(y_test, y_pred)
        plt.plot(recall, precision, color=colors.pop(), lw=0.5, label=f'{name} (F1 = {f1score:.4f})')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.xlim(-.01, 1.01)
        plt.ylim(-.01, 1.01)
        plt.title('Precision-Recall curve')
        plt.legend(loc='lower right')
        plt.grid(True)
```

```python
def plot_feature_importance(model):
    feature_cols = [x for x in X_train.columns if x != 0]
    feature_imp = pd.Series(model.feature_importances_, index=feature_cols).sort_values(ascending=False)

    ax = feature_imp.plot(kind='bar')
    ax.set(ylabel='Relative Importance')
```
